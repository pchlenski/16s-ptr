# Does greengenes have multiple sequences per gene?
> If so, we can do this analysis for anything greengenes-annotated

```python
import pandas as pd
from collections import defaultdict
```

```python
seq_in_gene = defaultdict(dict)
gene_in_seq = defaultdict(dict)
seqs_per_gene = defaultdict(lambda: 0)
genes_per_seq = defaultdict(lambda: 0)

with open("/home/phil/aptr/data/gg_99_otu_map.txt", "r") as f:
    for line in f.readlines():
        line_split = line.strip().split("\t")
        gene = line_split[0]
        seqs = line_split[1:]
        for seq in seqs:
            seq_in_gene[seq][gene] = 1
            gene_in_seq[gene][seq] = 1
            seqs_per_gene[gene] += 1
            genes_per_seq[seq] += 1
```

```python
for seq in genes_per_seq:
    if genes_per_seq[seq] > 1:
        print(seq, genes_per_seq[seq])
```
